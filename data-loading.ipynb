## Loading data

```python
import pandas as pd
import numpy as np
```

```python
# Display the dataframe
pd.set_option('display.max_columns', None)  # or 1000
pd.set_option('display.max_rows', None)  # or 1000
pd.set_option('display.max_colwidth', None)  # or 199
```

```python
# Read the data
X = pd.read_csv('data/train.csv', index_col='Id')

# Remove rows with missing target, separate target from predictors
X.dropna(axis=0, subset=['SalePrice'], inplace=True)

# Predictor variable y
y = X.SalePrice

# Input variables
X.drop(['SalePrice'], axis=1, inplace=True)
```

### Parse date

```python
# start_time           | end_time         
# 2019-03-29 21:23:12  | 2019-03-30 09:11:13

dt_cols = ['start_time', 'end_time']
df = pd.read_excel('survey.xlsx', parse_date=dt_cols) # turning string into datetime64

# another way: using pd.to_datetime()
format_string = "%m%s%Y %H:%M:%S"
df['end_time'] = pd.to_datetime(df['end_time'], format=format_string)
```
